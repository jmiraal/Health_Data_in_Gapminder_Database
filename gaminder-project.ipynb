This is the description of the values according to the **gapminder** page.

    
- **year:** The year in which the values were taken.


- **country:** The country for each value.


- **hdi:** Human development index. Is an index used to rank the countries by level of 'human development'. It contains
    three dimnensions: health level, educational level, and living standard.

- **life_expectancy:** The average number of years a newborn child would live if current mortality patterns were to stay the     same.

- **gov_health_spend:** The average health expenditure per person that has been paid by gobernment entities, expressed in US     dollars using the exchange rate.

- **total_health_spend:** The average health expenditure per person, expressed in US dollars using the average exchange         rate.

- **private_health_spend:** Private share of health expenditure (%). Percentage of total health expenditure that has been       paid by private entities, for exampel commercial insurance and out-of-pocket payments.

### Merging The Data


* In this part all the  dataframes will be merged in a sigle dataframe called **df_global**. We use a dictionary to make it more easy and to use the key to name the columns in the new dataframe.


```python
#It creates dictionary of dataframes. The keys will be used to name the columns in the global datafrme.

dfs = {'hdi': df_hdi, 
       'life_expectancy': df_life_expectancy, 
       'gov_health_spend': df_gov_health_spend, 
       'total_health_spend': df_total_health_spend, 
       'private_health_spend_pc': df_private_health_spend_pc,
       'breast_cancer_deaths_w': df_breast_cancer_deaths_w, 
       'cervical_cancer_deaths_w': df_cervical_cancer_deaths_w, 
       'colonandrectum_cancer_deaths_w': df_colonandrectum_cancer_deaths_w, 
       'liver_cancer_deaths_w': df_liver_cancer_deaths_w, 
       'lung_cancer_deaths_w': df_lung_cancer_deaths_w,
       'stomach_cancer_deaths_w': df_stomach_cancer_deaths_w, 
       'colonandrectum_cancer_deaths_m': df_colonandrectum_cancer_deaths_m,
       'liver_cancer_deaths_m': df_liver_cancer_deaths_m, 
       'lung_cancer_deaths_m': df_lung_cancer_deaths_m,
       'prostate_cancer_deaths_m': df_prostate_cancer_deaths_m, 
       'stomach_cancer_deaths_m': df_stomach_cancer_deaths_m,
       'breast_cancer_new_cases_w': df_breast_cancer_new_cases_w, 
       'cervical_cancer_new_cases_w': df_cervical_cancer_new_cases_w,
       'colonandrectum_cancer_new_cases_w': df_colonandrectum_cancer_new_cases_w, 
       'liver_cancer_new_cases_w': df_liver_cancer_new_cases_w,
       'lung_cancer_new_cases_w': df_lung_cancer_new_cases_w, 
       'stomach_cancer_new_cases_w': df_stomach_cancer_new_cases_w,
       'colonandrectum_cancer_new_cases_m': df_colonandrectum_cancer_new_cases_m, 
       'liver_cancer_new_cases_m': df_liver_cancer_new_cases_m,
       'lung_cancer_new_cases_m': df_lung_cancer_new_cases_m, 
       'prostate_cancer_new_cases_m': df_prostate_cancer_new_cases_m,
       'stomach_cancer_new_cases_m': df_stomach_cancer_deaths_m, 
      }

#It changes the structure of the original dataframes. 
#We create a list of dataframe, each one with the columns 'country', 'year' and 'value'.

dfs1 = [dfs[key].melt(id_vars=["country"], var_name = "year", value_name = key) for key in dfs]

#It performs a merge of all the data frames included in the previous list.

df_global = reduce(lambda left,right: pd.merge(left,right,on=['country', 'year'],
                                               how='outer'), dfs1).sort_values(by=['country', 'year'])

#It performs a separeted merge with the dataframe regions. That is because regions doesn't have the column year. 
#The countries are grouped in differente regions.

df_global = df_global.merge(df_regions, left_on=['country'], right_on=['country'], how='outer')
```

<a id='add_columns'></a>
### Adding New Columns


* It adds some additional columns calculated from the original data. For example we calculate the private health expenditure per person in US dollars using the percent value in **private_health_spend_pc** column.
  It also adds columns with the **totals of cancer deaths and cases**, summing the corresponding columns.


```python
#It calculates the value of total dollars spended in private health for 100K habitants 
#using the percentage value 'private_health_spend_pc'

df_global['private_health_spend'] = round(df_global['private_health_spend_pc'] * df_global['total_health_spend'] / 100, 2)

#It adds a column with the total cancer deaths for men.

cancer_men_list_death = ['colonandrectum_cancer_deaths_m', 
                         'liver_cancer_deaths_m', 
                         'lung_cancer_deaths_m', 
                         'prostate_cancer_deaths_m', 
                         'stomach_cancer_deaths_m']

df_global['total_cancer_men_death'] = df_global[cancer_men_list_death].sum(axis=1)

#It adds a column with the total cancer deaths for women.

cancer_women_list_death = ['breast_cancer_deaths_w', 
                           'cervical_cancer_deaths_w', 
                           'colonandrectum_cancer_deaths_w', 
                           'liver_cancer_deaths_w', 
                           'lung_cancer_deaths_w',
                           'stomach_cancer_deaths_w']

df_global['total_cancer_women_death'] = df_global[cancer_women_list_death].sum(axis=1)

#It adds a column with the total cancer cases for men.

cancer_men_list_cases = ['colonandrectum_cancer_new_cases_m', 
                         'liver_cancer_new_cases_m', 
                         'lung_cancer_new_cases_m', 
                         'prostate_cancer_new_cases_m', 
                         'stomach_cancer_new_cases_m']

df_global['total_cancer_men_cases'] = df_global[cancer_men_list_cases].sum(axis=1)

#It adds a column with the total cancer cases for women.

cancer_women_list_cases = ['breast_cancer_new_cases_w', 
                           'cervical_cancer_new_cases_w', 
                           'colonandrectum_cancer_new_cases_w', 
                           'liver_cancer_new_cases_w', 
                           'lung_cancer_new_cases_w',
                           'stomach_cancer_new_cases_w']

df_global['total_cancer_women_cases'] = df_global[cancer_women_list_cases].sum(axis=1)

#It sums both cases for women and men.

df_global['total_cancer_cases'] = df_global['total_cancer_women_cases'] + df_global['total_cancer_men_cases']
df_global['total_cancer_death'] = df_global['total_cancer_women_death'] + df_global['total_cancer_men_death']

#It calculates year by year the difference and percentage betwen total new cases and deaths. It gives us an estimation of the
#treatments against cancer success.

df_global['total_cancer_heal'] = df_global['total_cancer_cases'] - df_global['total_cancer_death']
df_global['total_cancer_heal_pc'] = round((df_global['total_cancer_cases'] - df_global['total_cancer_death']) 
                                          / df_global['total_cancer_cases'] * 100, 2)
```

### Research Question 1: What Is The Influence Of The Ratio Private/Total Health Expenditure In The Life Expectancy And In The Cancer Treatments Success?

* In this first question we are going to explore the differences between the public and private health investments. We are interested in knowing how relevant are they in reaching a high life expectancy and in the succeess of the treatments against cancer. Are more successful those countries with a higher public spending or those with a higher private health spending? 

  To make this aproach we are going to start depicting a graph with the percentage of the private health spending over the total spending based in the years of life expectancy. In the x-axis we are going to divide the countries by life expectancy years in four quartiles. The first quartile represents the countries with smallest life expectancy and the last quartile corresponds with the countries with a highest number of years. We are going to use the dataframe **df_health1** filtered for the year 2010.
  
  The table and the graph with the results are showed below:



```python
#We make a copy of df_health1 filtered by year 2010
df1 = df_health1.copy()
df1 = df1.query('year == "2010"')
#It calculates 4 quartiles for the life expectancy.
df1['Life Expectancy Quartile'] = pd.qcut(df1.life_expectancy, 
                                          4, 
                                          labels=['First Quartile', 'Second Quartile', 'Third Quartile', 'Forth Quartile'])
#rename the columns so that the columns look better
df1.rename(columns=lambda x: x.strip().title().replace("_", " "), inplace=True)
#Group by quartiles
df1 = df1.groupby('Life Expectancy Quartile').mean().round(2)
df1[['Life Expectancy', 'Private Health Spend Pc']]
```

```python
#It draws the result in a bar plot.
plt.subplots(figsize=(9, 6))
plt.bar(df1.index, df1['Private Health Spend Pc'])
plt.title('Private Health/Total Health Spending (%) by Life Expectency Quartiles', 
          fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.xlabel('Life Expectancy Quartiles (Years)')
plt.ylabel('(%)')
plt.xticks(fontsize=10, rotation='40')
plt.xticks(np.arange(4), df1['Life Expectancy']);
```

We can see that the ratio is smaller for the countries with a higher life expectancy. In these countries, with almost 80 years of life expectancy, the private health represents the 30% of the total health spending, while in the first quartile represents the 50%

* As we said before, we are also interested in knowing how does the ratio between private/total(%) health investment influence in the ratio of deaths versus total the cases of cancer. The calculations are made also for the year 2010.

```python
#We make a copy of df_health1 filtered by year 2010
df2 = df_health1.copy()
df2 = df2.query('year == "2010"')
#It calculates 4 quartiles for the Total Cancer Healing Percentage.
df2['Heal Percent Quartile'] = pd.qcut(df2.total_cancer_heal_pc, 
                                       4, 
                                       labels=['First Quartile', 'Second Quartile', 'Third Quartile', 'Forth Quartile'])
#rename the columns so that the columns look better
df2.rename(columns=lambda x: x.strip().title().replace("_", " "), inplace=True)
#Group by quartiles
df2 = df2.groupby('Heal Percent Quartile').mean().round(2)
df2[['Total Cancer Heal Pc', 'Private Health Spend Pc']]
```

### Research Question 2: Compare The Increments In Total Health Spending Between 1995 and 2010 With The Increments In The Life Expectancy And Healing Cancer Ratios During These Years.



* This question is related to the previous one, but in this case we want to put the focus in the total health spending and its growth between 1995 and 2010 for each region. We also want to compare this increments with the achievements in terms of reaching a higher life expectancy  and a cancer healing rates increments. The analisys will be made dividing the world in nine regions.

  Finally we will try to estimate how much it costs to increase a year in the life expectancy acording the current life expectancy. Or, which is the same, how much costs, in mean, to increase the life expectancy from 55 years to 56, from 65 to 66, 75 to 76, etc.

```python
#First we filter the dataframe df_health1 for 2010 and we group it by regions, 
#calculating the mean for each region.
df_10 = df_health1.copy().query('year == "2010"')
df_10_regions = df_10.groupby('nine_regions').mean()
df_10.info()
```

```python
#we do de same for 1995
df_95 = df_health1.copy().query('year == "1995"')
df_95_regions = df_95.groupby('nine_regions').mean()
df_95.info()
```

```python
#we sustract the previous dataframes to see the increments by region
df_inc_regions = df_10_regions - df_95_regions
#We drop the columns that we don't need now.
df_inc_regions.drop(columns = ['gov_health_spend', 'private_health_spend', 'private_health_spend_pc'],inplace = True)
#Change the name of the columns.
df_inc_regions.rename(columns=lambda x: x.strip().title().replace("_", " "), inplace=True)
df_inc_regions.rename(index=lambda x: x.strip().title().replace("_", " "), inplace=True)
df_inc_regions.rename(columns = {'Life Expectancy': 'Life Expectancy Inc', 
                       'Total Cancer Heal Pc': 'Total Cancer Heal Pc Inc', 
                       'Total Health Spend': 'Total Health Spend Inc'}, inplace = True)
df_inc_regions.round(2)
```

In this plot is showed that the region of the world tha has incremented most the life expectancy is West Asia wiht 3.97 years, while Central America only has had an increment of 2.54 years between 1995 and 2010.

But, in terms of health investments, North America and West Europe are by far the two regions which have increased more their spendings.

* Finally, we would like to estimate how much needs to increase its health investment a country in relation with its life expectancy. i.e. how much more expensive is to increase a year if its life expectancy is 75 than if its life expectancy is 55, for example.

```python
#rename the columns in both dataframes for both years before merging them.
df_10_mod = df_10.rename(columns=lambda x: x + "_2010")
df_95_mod = df_95.rename(columns=lambda x: x + "_1995")
```

```python
#It merges the dataframes for years 1995 and 2010.
df_combined = df_10_mod.merge(df_95_mod, left_on='country_2010', right_on='country_1995', how='inner')
#Adds new columns with the increment in the health spending and life expenctancy.
df_combined['total_health_spend_inc'] = df_combined['total_health_spend_2010'] - df_combined['total_health_spend_1995']
df_combined['life_expectancy_inc'] = df_combined['life_expectancy_2010'] - df_combined['life_expectancy_1995']
#We select the countries where life expectancy has grown
df_combined = df_combined[df_combined['life_expectancy_inc'] > 0] 
#For each country we divide the increment in the health spending by de years of increment.
df_combined['total_health_spend_per_year'] = round(df_combined['total_health_spend_inc'] / df_combined['life_expectancy_inc'], 2)
#We make groups of 5 years in the life expectancy column.
df_combined['age_groups'] = (df_combined['life_expectancy_1995'] / 5).astype(int) * 5
#Rename all the columns to have a better view of the table.
df_combined.rename(columns=lambda x: x.strip().title().replace("_", " "), inplace=True)
#It calcs the mean for each group.
df_combined = df_combined.groupby('Age Groups').mean().round(2)
df_combined
```

```python
#Draw the result in a bar plot. The x-axis represents groups of five years in the life expectancy.
#The y-axis represent how much it costs(the mean) to increase a year in the life expectancy for each group.
plt.subplots(figsize=(9, 6))
plt.bar(df_combined.index, df_combined['Total Health Spend Per Year'], width = 3.5)
plt.title('Cost of Incrementing a Year in Life Expectancy', fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.xlabel('Life Expectancy in 1995')
plt.ylabel('$/Year Increm./100K')
plt.xticks(fontsize=10, rotation='vertical');
```

This graph shows that around 75 years the investments in health care of countries per incremented year are much bigger that for countries with lower life expectancy. This is expectable taking into account that these countries are closer to the limit and they are usually the richest ones.

### Research Question 3: Relation Between The Cancer Types And The Region Of The World Or The HDI (Human Development Index).

* In this part we are interested in knowing if there are some correlation between the HDI or the region of the world and the different cancer types.

  To analyze that we are going to investigate which is the most usual cause of cancer death in each region or in each group of HDI. The calculations will be made for 1995 and 2015. That will permit us to find out any difference between regions and see if there has been any change with time.

  We are going to start with the analysis by the Human Development Index (HDI). We separate the countries in quartiles again by the HDI. Then, it is calculated the mean of deaths in each quartile for each type of cancer. We select the causes with more deaths in each quartile and it is showed in a dataframe and in a bar plot. The analisys will be mado for both years.

```python
#Filter for year 1995. Group by quartiles of Human Development Index (hdi)
df5 = df_health2.query('year == "1995"')
df5 = df5.drop(columns = ['nine_regions'])
df5.hdi = pd.qcut(df5.hdi, 4, labels=['hdi25', 'hdi50', 'hdi75', 'hdi100'])
df5 = df5.groupby('hdi', as_index = False).mean().round(2)

#Filter for year 2015. Group by quartiles of Human Development Index (hdi)
df6 = df_health2.query('year == "2015"')
df6 = df6.drop(columns = ['nine_regions'])
df6.hdi = pd.qcut(df6.hdi, 4, labels=['hdi25', 'hdi50', 'hdi75', 'hdi100'])
df6 = df6.groupby('hdi', as_index = False).mean().round(2)
cols = cancer_men_list_death + cancer_women_list_death

#Get the cause of cancer with most deaths. Save the result for the two years in a single dataframe.
df5['max_deaths_1995'] = df5[cols].max(1)
df5['max_deaths_2015'] = df6[cols].max(1)
df5['max_cancer_case_1995'] = df5[cols].idxmax(1).str.title().str.replace("_", " ").str.replace('Deaths', '')
df5['max_cancer_case_2015'] = df6[cols].idxmax(1).str.title().str.replace("_", " ").str.replace('Deaths', '')
df5 = df5[['hdi', 'max_deaths_1995', 'max_cancer_case_1995', 'max_deaths_2015', 'max_cancer_case_2015']]
df5.rename(columns=lambda x: x.strip().title().replace("_", " "), inplace=True)
df5.set_index('Hdi')
```

```python
#It draws the result in a bar plot.
#In the x-axis we set the four quarties of HDI
#In the y-axis we can see the mean of number of deaths for 100K people for the max cause of deaths.
#The results for the two years are showed in the same plot.
plt.subplots(figsize=(9, 6))
x = [0, 1, 2, 3]
x1 = [0.4, 1.4, 2.4, 3.4]
plt.bar(x, df5['Max Deaths 1995'], width=0.4)
plt.bar(x1, df5['Max Deaths 2015'], width=0.4)
plt.title('Maximum Type of Cancer by HDI', fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.xlabel('HDI Quartile')
plt.ylabel('Total Deaths/100K')
plt.xticks(fontsize=11, rotation=10);
plt.yticks(fontsize=11)
#We add the name of the cause of cancer on each bar.
for i, v in enumerate(df5['Max Cancer Case 1995']):
    plt.text(i, 5, str(v), rotation=90, fontsize=10)
for i, v in enumerate(df5['Max Cancer Case 2015']):
    plt.text(i+0.4, 5, str(v), rotation=90, fontsize=10)
plt.xticks(np.arange(4)+0.2, ['hdi25%', 'hdi50%', 'hdi75%', 'hdi100%'])
plt.legend(['1995','2015'],loc=2, fontsize = 11);
```

We have found out some differences. In all the countries with high development index, the lung cancer in men seems to be the more usual and there has not been changes in the last years, except a decrement in the number of deaths by 100K people.

On the other hand, in the low developed countries, the principal causse of cancer in 2015 seems to be the prostate cancer. And some years ago the most usual cause of death in the lowest developed countries was the Cervical Cancer in womans.

There are also some types of cancer that have not almost correlation.
